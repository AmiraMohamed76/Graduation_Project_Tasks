```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt   
```

```python
image = cv2.imread('/home/linux/Desktop/images (8).jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
hand_cascade = cv2.CascadeClassifier('/home/linux/Desktop/Hand.Cascade.1.xml')
```

```python
def detect_hand(img):
    
  
    img_copy = img.copy()
  
    hand_rects = hand_cascade.detectMultiScale(img_copy)
    
    for (x,y,w,h) in hand_rects: 
        cv2.rectangle(img_copy, (x,y), (x+w,y+h), (255,0,0), 15) 
        
    return img_copy
```

```python
result = detect_hand(image)
```

```python
plt.imshow(result,cmap='gray')
```
